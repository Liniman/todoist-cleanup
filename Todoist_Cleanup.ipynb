```python
!pip install requests ipywidgets

import requests
import ipywidgets as widgets
from IPython.display import display

# Set up your API token
API_TOKEN = ""

# Function to get all tasks
def get_all_tasks():
    url = "https://api.todoist.com/rest/v2/tasks"
    headers = {
        "Authorization": f"Bearer {API_TOKEN}"
    }
    response = requests.get(url, headers=headers)
    return response.json()

# Function to delete a task
def delete_task(task_id):
    url = f"https://api.todoist.com/rest/v2/tasks/{task_id}"
    headers = {
        "Authorization": f"Bearer {API_TOKEN}"
    }
    response = requests.delete(url, headers=headers)
    return response.status_code

# Function to get all labels
def get_all_labels():
    url = "https://api.todoist.com/rest/v2/labels"
    headers = {
        "Authorization": f"Bearer {API_TOKEN}"
    }
    response = requests.get(url, headers=headers)
    return response.json()

# Function to delete a label
def delete_label(label_id):
    url = f"https://api.todoist.com/rest/v2/labels/{label_id}"
    headers = {
        "Authorization": f"Bearer {API_TOKEN}"
    }
    response = requests.delete(url, headers=headers)
    return response.status_code

# Function to display tasks with checkboxes
def display_tasks_with_checkboxes():
    tasks = get_all_tasks()
    task_checkboxes = []

    for task in tasks:
        checkbox = widgets.Checkbox(value=True, description=task['content'])
        task_checkboxes.append((checkbox, task['id']))

    return task_checkboxes

# Function to display labels with checkboxes
def display_labels_with_checkboxes():
    labels = get_all_labels()
    label_checkboxes = []

    for label in labels:
        checkbox = widgets.Checkbox(value=True, description=label['name'])
        label_checkboxes.append((checkbox, label['id']))

    return label_checkboxes

# Display tasks
task_checkboxes = display_tasks_with_checkboxes()
print("Tasks:")
for checkbox, _ in task_checkboxes:
    display(checkbox)

# Display labels
label_checkboxes = display_labels_with_checkboxes()
print("\nLabels:")
for checkbox, _ in label_checkboxes:
    display(checkbox)

# Function to delete everything except checked tasks and labels
def delete_all_except_selected(button):
    for checkbox, task_id in task_checkboxes:
        if not checkbox.value:
            print(f"Kept task: {checkbox.description}")
        else:
            delete_task(task_id)
            print(f"Deleted task: {checkbox.description}")

    for checkbox, label_id in label_checkboxes:
        if not checkbox.value:
            print(f"Kept label: {checkbox.description}")
        else:
            delete_label(label_id)
            print(f"Deleted label: {checkbox.description}")

# Create a continue button
continue_button = widgets.Button(description="Continue")
continue_button.on_click(delete_all_except_selected)

# Display the continue button
display(continue_button)
```
